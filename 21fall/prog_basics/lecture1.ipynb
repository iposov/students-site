### История

Задание. Почитать википедию про Python.

Когда-то у языка была первая версия, потом вышла
вторая. Python 2, улучшалась, выходили версии
Python 2.1, 2.2, ...
В какой-то момент стало понятно, что необходимы
улучшения языка, которые нарушат *обратную
совместимость*, т.е. после введения улучшений
окажется, что невозможно запускать старые
программы. Например,

в Python2 писали `print "Hello!"`. Не нужны
были скобки в `print`, но в других ситуациях
скобки ставили: `show("Hello")`, `sin(pi)`,
`f(x)`. Улучшение, сделать так, что скобки
нужны везде. Старая программка теперь без
скобок не запустится. Еще пример, работа с
текстом. В Python2 строки были без кодировок,
программист сам должен был знать, в какой
они кодировке. Хочется, чтобы кодировка была
единообразная, utf-8. Много других примеров.

Была выпущена версия Python 3, которая
несовместима с Python 2. На Python 2 уже было
очень много написано кода, поэтому невозможно
было сразу всем перейти на версию 3. До сих
пор есть очень много кода на Python 2.
Если будете искать библиотеки, примеры кода,
следите, что они подходят для Python 3.

Сейчас последние версии - это Python 2.7 и
Python 3.9. Версия 3 продолжает развиваться дальше,
версия 2 больше не поддерживается.

### Установка, запуск программ, первая программа

Загрузить можно по ссылке https://www.python.org.
Это официальный дистрибутив Python. На самом деле
есть другие, их много, есть, например, реализация
на Java. Есть эффективные реализации, есть
дистрибутивы со специализированными пакетными
менеджерами, например, Anaconda.

Нужно скачать, установить дистрибутив, вы получите
каталог с питоном, там будут:
  * интерпретатор python.exe
  * среду разработки IDLE

Попробуем сначала интерпретатор python.exe:
Если просто запустить python.exe, получем черную
консоль интерпретатора, в которой можно вводить
команды на python:

```python
2 + 2
```

```python
print("Hello")
```

```python
2 * 100
```

```python
2 ** 100
```

Пользуйтесь обязательно python-консолью для изучения
языка, вы можете ввести любое выражение и посмотреть,
как оно вычисляется.

Кроме этого, python.exe может запускать программы
из файла. (проделали вживую)

Сделаем то же самое в среде разработки IDLE.
В ней есть консоль и текстовый редактор для
написания кода.
